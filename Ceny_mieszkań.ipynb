```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Heating'] = le.fit_transform(df['Heating'])
df['Parking_Space'] = le.fit_transform(df['Parking_Space'])
df['Balcony_Garden_Terrace'] = le.fit_transform(df['Balcony_Garden_Terrace'])
df['Finishing_Condition'] = le.fit_transform(df['Finishing_Condition'])
df.head()
```

```python
df.drop(df.loc[df['Number_of_Rooms'] == 'więcej niż 10'].index, inplace=True)
```

```python
df.drop('City', axis = 1, inplace = True)
```

```python
voivodeship_groupby_mean = df.groupby('Voivodeship')['Price'].mean()
```

```python
df['mean_price_in_voivodeship'] = df['Voivodeship'].map(voivodeship_groupby_mean).astype(int)
```

```python
df.head()
```

```python
df['mean_m2_price'] = df['Price'] / df['Surface']
df['mean_m2_price'] = df['mean_m2_price'].astype(int)
df.head()
```

```python
voivodeship_groupby_m2_mean = df.groupby('Voivodeship')['mean_m2_price'].mean()
df['mean_m2_price_in_voivodeship'] = df['Voivodeship'].map(voivodeship_groupby_m2_mean).astype(int)
df.head()
```

```python
def price_category(row):

  voivodeship = row['Voivodeship']
  mean_price = row['mean_m2_price']

  mean_price_m2_in_voivodeship = df[df['Voivodeship'] == voivodeship]['mean_m2_price_in_voivodeship'].mean()

  if mean_price < mean_price_m2_in_voivodeship * 0.3:
    return 'cheap'
  elif mean_price < mean_price_m2_in_voivodeship * 0.8:
    return 'normal'
  else:
    return 'expensive'
```

```python
df['price_category'] = df.apply(price_category, axis=1)
df.head()
```

```python
df['price_category'].value_counts()
```

```python
data = df[['Voivodeship','mean_m2_price','Number_of_Rooms','Parking_Space']].copy()
target = df['price_category'].copy()
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Voivodeship'] = le.fit_transform(data['Voivodeship'])
target = le.fit_transform(target)

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=42)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_combined_std = np.vstack((X_train_scaled, X_test_scaled))
y_combined = np.hstack((y_train, y_test))





```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print('Random Forest Accuracy for n_estimators \n')
for i in np.arange(1 , 250, 25 -1):
  rf_classifier = RandomForestClassifier(n_estimators=i, random_state=42)
  rf_classifier.fit(X_train_scaled, y_train)
  y_pred = rf_classifier.predict(X_test_scaled)
  accuracy = accuracy_score(y_test, y_pred)
  print(f"Accuracy for {i} estimators: {accuracy}")

from sklearn.svm import SVC

print('\n SVC accuracy for C = i \n')

for i in range(1,11):
  svm = SVC(kernel='rbf', C=i)
  svm.fit(X_train_scaled, y_train)
  y_pred = svm.predict(X_test_scaled)
  accuracy = accuracy_score(y_test, y_pred)
  print(f"Accuracy for {i} estimators: {accuracy}")

from sklearn.linear_model import LogisticRegression

print('\n')
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)
y_pred = lr.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy for Logistic Regression: {accuracy}")
```

```python
rf_classifier = RandomForestClassifier(n_estimators=1, random_state=42)
rf_classifier.fit(X_train_scaled, y_train)
y_pred = rf_classifier.predict(X_test_scaled)

accuracy_score = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy_score}")
```
